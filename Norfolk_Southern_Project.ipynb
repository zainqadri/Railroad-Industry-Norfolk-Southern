```python
#Show the dataframe in descending order by Report Year

trimmeddf_date = trimmeddf.sort_values(by=['Report Year'], ascending=False)
trimmeddf_date
```

```python
#check if the dataframe is indexed or not
print(trimmeddf_date.index)
```

```python
#Drop the row without any data in it
trimmeddf_date = trimmeddf_date.drop(labels=71767, axis = 0, inplace = False)
trimmeddf_date
```

```python
#calculate the average to find the industry average for Employees on Duty during an Accident
averagedf = trimmeddf_date.groupby(['Report Year']).mean()
averagedf
```

## Data Visualization to Show the Number of Employees on Duty during an Accidnet for the Railroad Industry.
### The number of employees has decreased overall

```python
#Create Data Visualizations To Show the Industry Average of Number of Employees during a reported accident 
averagedf.reset_index().plot.line(x='Report Year', y='Engineers On Duty', color='Green', alpha=0.9, linewidth= 2)
font1 = {'family':'Serif','color':'Black','size':13}
font2 = {'family':'Serif','color':'Black','size':10}
plt.title('Average Number of Engineers by Year (Industry)', fontdict = font1, pad = 13.0)
plt.xlabel('Year', fontdict = font2, labelpad=6.0)
plt.ylabel('Number of Employees', fontdict = font2, labelpad=8.0)
plt.show()

averagedf.reset_index().plot(x='Report Year', y='Firemen On Duty', color='Green', alpha=0.9, linewidth=2)
plt.title('Average Number of Fireman by Year (Industry)', fontdict = font1, pad = 13.0)
plt.xlabel('Year', fontdict = font2, labelpad=6.0)
plt.ylabel('Number of Employees', fontdict = font2, labelpad=8.0)
plt.show()

averagedf.reset_index().plot(x='Report Year', y='Conductors On Duty', color='Green', alpha=0.9, linewidth=2)
plt.title('Average Number of Conductors by Year (Industry)', fontdict = font1, pad = 13.0)
plt.xlabel('Year', fontdict = font2, labelpad=6.0)
plt.ylabel('Number of Employees', fontdict = font2, labelpad=8.0)
plt.show()

averagedf.reset_index().plot(x='Report Year', y='Brakemen On Duty', color='Green', alpha=0.9, linewidth=2)
plt.title('Average Number of Brakemen by Year (Industry)', fontdict = font1, pad = 13.0)
plt.xlabel('Year', fontdict = font2, labelpad=6.0)
plt.ylabel('Number of Employees', fontdict = font2, labelpad=8.0)
plt.show()
```

```python
#Find the Number of Accidents in the Industry Per Year
countdf = trimmeddf_date.groupby(['Report Year']).count()
countdf.rename(columns={"Reporting Railroad Code": "Number of Accidents"}, inplace = True)
countdf = countdf.filter(items=['Number of Accidents'])
countdf
```

## Data Visualization to Show the Number of Accidents in the Railroad Industry
### The number of accidents has decreased overall

```python
#Create Data Visualizations To Show the Number of Accidents Per Year in the Railroad Industry
countdf.reset_index().plot.line(x='Report Year', y='Number of Accidents', color='Green', alpha=0.9, linewidth= 2)
font1 = {'family':'Serif','color':'Black','size':13}
font2 = {'family':'Serif','color':'Black','size':10}
plt.title('Number of Accidents by Year (Industry)', fontdict = font1, pad = 13.0)
plt.xlabel('Year', fontdict = font2, labelpad=6.0)
plt.ylabel('Number of Accidents', fontdict = font2, labelpad=8.0)
plt.show()
```

```python
#Add up All the Saftey Employees on Hand during accidents, and create a new column to Countdf
countdf["Sum of Average Number of Saftey Employees on Duty"] = averagedf["Engineers On Duty"] + averagedf["Firemen On Duty"] + averagedf["Conductors On Duty"] + averagedf["Brakemen On Duty"]
countdf
```
